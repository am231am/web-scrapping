## scrape jumia

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import requests
import urllib
import time
import pandas as pd
```

```python
driver=webdriver.Chrome()
```

```python
page="https://www.jumia.com.eg/catalog/?q=phone"
driver.get(page)
```

```python
product_info = driver.find_elements(By.XPATH, "//article[@class='prd _fb col c-prd']")
```

```python
# scroll dowen
Height=0
for i in range(12):
    Height=Height+480
    driver.execute_script(f"window.scrollTo(0,{Height})")
    time.sleep(1)
```

```python
all_pages=[]
for i in range(1,51):
    pages=page+"&page="+str(i)
    all_pages.append(pages)   
```

```python
product_name = []
product_prices = []
product_rate = []
product_image = []
product_image_link=[]

for page in all_pages:
    for info in product_info:
        name=info.find_element(By.XPATH,".//a[@class='core']/div[@class='info']/h3[@class='name']").text
        
        try:
            price = info.find_element(By.XPATH, ".//a[@class='core']/div[@class='info']/div[@class='prc']").text
        except NoSuchElementException:
            price = "none"
        try:
            rate = info.find_element(By.XPATH, ".//a[@class='core']/div[@class='info']/div[@class='rev']/div").text
        except NoSuchElementException:
            rate = "none"
        
        img_link=info.find_element(By.XPATH,".//a[@class='core']/div[@class='img-c']/img").get_attribute('src')
        img=info.find_element(By.XPATH,".//a[@class='core']/div[@class='img-c']/img")
        
        if price != "none" and rate != "none":
            product_name.append(name)
            product_prices.append(price)
            product_rate.append(rate.split("o")[0])
            product_image_link.append(img_link)
            product_image.append(img)
```

```python
all_data={
    'product_name':product_name
    ,'product_prices':product_prices
    ,'product_rate':product_rate
    ,'product_image':product_image_link
}

df=pd.DataFrame(all_data)
df
```

```python
df.to_csv("phone.csv")
```
